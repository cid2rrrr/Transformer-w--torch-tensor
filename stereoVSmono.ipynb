```python
import librosa
import librosa.display
import scipy.io
```

```python
stereo, sr = librosa.load(r"C:\Users\KIST\Downloads\drums.mp3", mono=False)
monoL = stereo[0]
```

```python
monoLR, sr = librosa.load(r"C:\Users\KIST\Downloads\drums.mp3", mono=True)
```

```python
L_onset_frames = librosa.onset.onset_detect(y=monoL, sr=sr
            , pre_max=0, post_max=2
            , pre_avg=0, post_avg=2
            , delta=0.02
            , wait=2
            )
L_onset_times = librosa.frames_to_time(L_onset_frames)
```

```python
LR_onset_frames = librosa.onset.onset_detect(y=monoLR, sr=sr
            , pre_max=0, post_max=2
            , pre_avg=0, post_avg=2
            , delta=0.02
            , wait=2
            )
LR_onset_times = librosa.frames_to_time(LR_onset_frames)
```

```python
L_clicks = librosa.clicks(frames=L_onset_frames, sr=sr, length=len(monoL))
LR_clicks = librosa.clicks(frames=LR_onset_frames, sr=sr, length=len(monoLR))
```

```python
scipy.io.wavfile.write("l_click.wav", 22050, L_clicks)
scipy.io.wavfile.write("lr_click.wav", 22050, LR_clicks)
```
