```python
import numpy as np
import matplotlib.pyplot as plt

from IPython.display import Audio

import librosa
import librosa.display
```

```python
y, sr = librosa.load(r"C:\Users\KIST\Downloads\One Eleven.wav")

Audio(data=y, rate=sr)
```

```python
D = librosa.stft(y)
```

```python
D_harmonic, D_percussive = librosa.decompose.hpss(D, margin=2)
```

```python
y_percussive = librosa.istft(D_percussive, length=len(y))
```

```python
D_harmonic8, D_percussive8 = librosa.decompose.hpss(D, margin=8)
y_percussive8 = librosa.istft(D_percussive8, length=len(y))
```

```python
import scipy.io


scipy.io.wavfile.write("./perc/margin2.wav", sr, y_percussive)
scipy.io.wavfile.write("./perc/margin8.wav", sr, y_percussive8)
```
